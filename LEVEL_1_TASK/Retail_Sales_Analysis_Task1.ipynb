```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the Dataset
# Ensure the CSV file is in the same folder as this notebook
df = pd.read_csv('retail_sales_dataset.csv')

# 2. Data Cleaning
print("Missing values per column:")
print(df.isnull().sum())

# 3. Descriptive Statistics
# This provides the Mean, Median, and Mode for the dataset
print("\nStatistical Summary:")
print(df.describe())

# 4. Time Series Analysis (Monthly Sales Trend)
# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Create a new column for Year-Month
df['Month'] = df['Date'].dt.to_period('M')

# Group by Month and sum the Total Amount
monthly_sales = df.groupby('Month')['Total Amount'].sum()

# Plotting the trend
plt.figure(figsize=(10, 5))
monthly_sales.plot(kind='line', marker='o', color='blue')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()

# 5. Customer & Product Analysis
# Revenue by Product Category
plt.figure(figsize=(10, 5))
sns.barplot(x='Product Category', y='Total Amount', data=df, estimator=sum, palette='viridis')
plt.title('Total Revenue by Product Category')
plt.show()

# 6. Revenue Distribution by Gender
plt.figure(figsize=(7, 7))
df.groupby('Gender')['Total Amount'].sum().plot(kind='pie', autopct='%1.1f%%', colors=['pink', 'lightblue'])
plt.title('Revenue Distribution by Gender')
plt.ylabel('')
plt.show()
```
